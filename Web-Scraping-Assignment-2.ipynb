# Q1: scrape data for “Data Analyst” Job position in “Bangalore” location job-title, job-location, company_name, experience_required.

```python
# by entering the below code url will be opened into the broswer.
driver.get("https://www.naukri.com/")
time.sleep(5)
## now lets enter the job title
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data analyst')
# enterning the job location 
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("bangalore")
time.sleep(5)
##using  class_name function for clicking 
search_btn=driver.find_element_by_xpath("//button[@class='btn']").click()
###extracting all job titles
job_titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
time.sleep(5)
companies_name=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
time.sleep(5)
# extracting all the companies name
company_names=[]
for i in companies_name:
       company_names.append(i.text) 
# extracting the tags having experience required data
Experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']//span")
Experience_list=[]
for i in Experience_tags:
    Experience_list.append(i.text)
time.sleep(5)
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span[1]")
location_list=[]
for i in location_tags:
    location_list.append(i.text)
location_list_1=[]
for i in range(0, len(location_list)):
    location_list_1.append(location_list[i])
time.sleep(5)

Data_Analyst = pd.DataFrame({})
Data_Analyst["Company"]=company_names[:10]
Data_Analyst["Experience"]=Experience_list[:10]
Data_Analyst["Location"]=location_list_1[:10]
Data_Analyst
```

# Q:2  scrape data for “Data Scientist” Job position in “Bangalore” location. scrape the job-title, job-location, company_name, full job-description.

```python
driver.get("https://www.naukri.com/")
## now lets enter the job title
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data Scientist')
# enterning the job location 
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("Bangalore")
time.sleep(5)
##using  class_name function for clicking 
search_btn=driver.find_element_by_xpath("//button[@class='btn']").click()
time.sleep(5)

jobtitle=[]
joblocation=[]
company_name=[]
full_jobdescription=[]

urls=[]
for i in driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']"):
    urls.append(i.get_attribute("href"))

for i in urls[:10]:
    driver.get(i)
    time.sleep(3)
    try:
        jobt=driver.find_element_by_xpath("//h1[@class='jd-header-title']")
        jobtitle.append(jobt.text.replace("/n","new line"))
        
        
    except:
        jobt=driver.find_element_by_xpath("//h1[@class='av-special-heading-tag ']")
        jobtitle.append(jobt.text.replace("/n","new line"))
        
    try:
        jobl=driver.find_element_by_xpath("//a[@title=' Jobs in Bangalore/Bengaluru']")
        joblocation.append(jobl.text)
        
        
    except:
        jobl=driver.find_element_by_xpath("//a[@class='cityLinks']")
        joblocation.append(jobl.text) 
        
    try:
        company_nam=driver.find_element_by_xpath("//div[@class='jd-header-comp-name']//a")
        company_name.append(company_nam.text)
        
        
    except:
        company_nam=driver.find_element_by_xpath("//p[@class='cpName f14']")
        company_name.append(company_nam.text)
                 
    try:
        full_jobdescriptio=driver.find_element_by_xpath("//section[@class='job-desc']")
        full_jobdescription.append(full_jobdescriptio.text)
        
        
    except:
        full_jobdescriptio=driver.find_element_by_xpath("//div[@class='clearboth description']")
        full_jobdescription.append(full_jobdescriptio.text)

Data_Scientist = pd.DataFrame({})
Data_Scientist["Jobtitle"]=jobtitle
Data_Scientist["Joblocation"]=joblocation
Data_Scientist["Company_name"]=company_name
Data_Scientist["Jobdescription"]=full_jobdescription
Data_Scientist
```

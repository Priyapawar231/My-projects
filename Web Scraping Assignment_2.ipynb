# Q1: scrape data for “Data Analyst” Job position in “Bangalore” location job-title, job-location, company_name, experience_required.

```python
##connecting the web driver 
driver=webdriver.Chrome(r"C:\Users\piyap\Downloads\chromedriver_win32\chromedriver.exe")
# by entering the below code url will be opened into the broswer.
url="https://www.naukri.com/"
driver.get(url)
time.sleep(5)
## now lets enter the job title
search_job = driver.find_element_by_id('qsb-keyword-sugg')
search_job.send_keys('Data analyst')
# enterning the job location 
search_loc=driver.find_element_by_id('qsb-location-sugg')
search_loc.send_keys("banglore")
time.sleep(5)
##using  class_name function for clicking 
search_btn=driver.find_element_by_xpath("//button[@class='btn']").click()
###extracting all job titles
job_titles=driver.find_elements_by_xpath("//a[@class='title fw500 ellipsis']")
time.sleep(5)
companies_name=driver.find_elements_by_xpath("//a[@class='subTitle ellipsis fleft']")
time.sleep(5)
# extracting all the companies name
company_names=[]
for i in companies_name:
       company_names.append(i.text) 
# extracting the tags having experience required data
Experience_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi experience']//span")
Experience_list=[]
for i in Experience_tags:
    Experience_list.append(i.text)
time.sleep(5)
location_tags=driver.find_elements_by_xpath("//li[@class='fleft grey-text br2 placeHolderLi location']//span[1]")
location_list=[]
for i in location_tags:
    location_list.append(i.text)
location_list_1=[]
for i in range(0, len(location_list)):
    location_list_1.append(location_list[i])
time.sleep(5)

Data_Analyst = pd.DataFrame({})
Data_Analyst["Company"]=company_names[:10]
Data_Analyst["Experience"]=Experience_list[:10]
Data_Analyst["Location"]=location_list_1[:10]
Data_Analyst
```
